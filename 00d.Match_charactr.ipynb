```python
import csv
import os
from os.path import isfile, join
from os import path
from pathlib import Path
import pandas as pd
```

```python
root = Path('/Users/saracaicedo/Library/CloudStorage/OneDrive-Universidaddelosandes/2025-1/Tesis')
csv_path = os.path.join(root,"input/database_CR/1979.csv")
cm_path = os.path.join(root,"input/congressman_data/congressman_1979.csv")
```

```python
original1979 = pd.read_csv(csv_path)
```

```python
titles = ['Mr.']

# Create new columns with default values
original1979['Prefix'] = None
original1979['Last name'] = None

# Split only rows that start with the specified titles
for title in titles:
    mask = original1979['speaker'].str.startswith(title)
    original1979.loc[mask, ['Prefix', 'Last name']] = (
        original1979.loc[mask, 'speaker'].str.split(' ', n=1, expand=True)
    )
```
